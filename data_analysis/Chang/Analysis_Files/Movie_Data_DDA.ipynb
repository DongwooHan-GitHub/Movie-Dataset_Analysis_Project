```python
import pandas as pd

import pandas as pd
import ast

# CSV 파일 경로
file_path_movies = '/content/drive/MyDrive/SeSAC/Mini_Project/movie_data/movies_metadata.csv'

# CSV 파일 읽기
df_movies = pd.read_csv(file_path_movies, low_memory=False)
df_movies.head()
```

```python
df_movies.info()
```

```python
df_movies_name_id = pd.read_csv(file_path_movies, usecols=['id','original_title'])
df = df_movies_name_id.drop_duplicates()
output_file_path = '/content/drive/MyDrive/SeSAC/Mini_Project/data/movie_data/movie_title_id.csv'
df.to_csv(output_file_path, index=False)
```

```python
file_path_movie_id = '/content/drive/MyDrive/SeSAC/Mini_Project/movie_data/movie_title_id.csv'
df_movie_id = pd.read_csv(file_path_movie_id, low_memory=False)
# 'id' 열을 숫자형으로 변환
df_movie_id['id'] = pd.to_numeric(df_movie_id['id'], errors='coerce').astype('Int64')
df = df_movie_id.sort_values(by='id', ascending=True)
# 인덱스를 재설정
df = df.reset_index(drop=True)
output_file_path = '/content/drive/MyDrive/SeSAC/Mini_Project/data/movie_data/movie_title_id_ver2.csv'
df.to_csv(output_file_path, index=False)
```
